```python
import pandas as pd

# Load dataset
df = pd.read_csv("unified_monthly_data_interpolated_1990_20250101.csv")

# Basic overview
print(df.info())
print(df.head())
```

```python
#chek duplicates 
duplicates = df.duplicated().sum()
df = df.drop_duplicates()

#check nuls
missing = df.isnull().sum()

```

```python
#conver Date to Datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#standardize
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['MedianSalesPriceofHousesSold', 'AverageSalesPrice']] = scaler.fit_transform(df[['MedianSalesPriceofHousesSold', 'AverageSalesPrice']])
#print(df.columns.tolist())

```

```python
#Outlier
from scipy import stats

z_scores = stats.zscore(df.select_dtypes(include='number'))
abs_z_scores = abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df = df[filtered_entries]
```

```python
#how many outliers were removed
removed_rows = df.shape[0] - filtered_entries.shape[0]
print("Number of rows removed due to outliers:", removed_rows)
```

```python
#Phase2 Part1 section2
import pandas as pd

# Load the dataset
housing_df = pd.read_csv('cleaned_housing_data.csv')

# Convert 'Date' to datetime format
housing_df['Date'] = pd.to_datetime(housing_df['Date'])

# Preview
housing_df.head()


```

```python
#Aggregation
# Example: Average housing price by Region
avg_price_by_province = housing_df.groupby('Region')['AverageSalesPrice'].mean().reset_index()
avg_price_by_province
```

```python
#Feature Engeneering
# Create a new feature: Price-to-Income Ratio
# Right now, we use MedianListingPriceperSquareFeet as a normalized price measure

housing_df['Price_per_SquareFoot'] = housing_df['MedianListingPriceperSquareFeet']

# Calculate supply-demand proxy
housing_df['Housing_Supply_Demand_Ratio'] = housing_df['MonthlySupplyofNewHouses'] / housing_df['NewOneFamilyHousesSold']

# Homeownership indicator
housing_df['Is_Homeownership_High'] = (housing_df['HomeownershipRate'] > 65).astype(int)
```

```python
# Extract year and month from the Date
housing_df['Year'] = housing_df['Date'].dt.year
housing_df['Month'] = housing_df['Date'].dt.month
```

```python
# Categorize MedianSalesPriceofHousesSold into affordability tiers
housing_df['Affordability_Level'] = pd.qcut(
    housing_df['MedianSalesPriceofHousesSold'],
    q=4,
    labels=['Most Affordable', 'Affordable', 'Expensive', 'Very Expensive']
)
```

```python
import pandas as pd
from scipy.stats import f_oneway, ttest_ind

# Load feature dataset
df = pd.read_csv('features_for_modeling.csv')

# One-way ANOVA: Is unemployment different across affordability categories?
anova_result = f_oneway(
    df[df['Affordability_Level'] == 'Most Affordable']['UnemploymentRate'],
    df[df['Affordability_Level'] == 'Affordable']['UnemploymentRate'],
    df[df['Affordability_Level'] == 'Expensive']['UnemploymentRate'],
    df[df['Affordability_Level'] == 'Very Expensive']['UnemploymentRate']
)
print("ANOVA for Unemployment Rate across Affordability Levels:", anova_result)

# t-test: HomeownershipRate - Most Affordable vs Very Expensive
t_test_result = ttest_ind(
    df[df['Affordability_Level'] == 'Most Affordable']['HomeownershipRate'],
    df[df['Affordability_Level'] == 'Very Expensive']['HomeownershipRate'],
    equal_var=False
)
print("T-test for HomeownershipRate between Most Affordable and Very Expensive:", t_test_result)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Load the feature dataset
df = pd.read_csv("features_for_modeling.csv")

# Separate features and target
X = df.drop(columns=["Affordability_Level"])
y = df["Affordability_Level"]

# Encode target labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)  # categorical: 0,1,2,3
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Train model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Predictions
y_pred = rf.predict(X_test)

# Report
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score

# 1. Load dataset
df = pd.read_csv("features_for_modeling.csv")

# 2. Prepare features and target
X = df.drop(columns=["Affordability_Level"])
y = df["Affordability_Level"]

# 3. Encode target labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)  # Convert categories to integers

# 4. Split data into train and test sets (stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# 5. Train Random Forest model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# 6. Train XGBoost model without use_label_encoder warning
xgb = XGBClassifier(eval_metric='mlogloss', random_state=42)
xgb.fit(X_train, y_train)

# 7. Make predictions
rf_pred = rf.predict(X_test)
xgb_pred = xgb.predict(X_test)

# 8. Print classification reports and accuracy scores
print("=== Random Forest Classification Report ===")
print(classification_report(y_test, rf_pred, target_names=label_encoder.classes_))
print("Random Forest Accuracy:", accuracy_score(y_test, rf_pred))

print("\n=== XGBoost Classification Report ===")
print(classification_report(y_test, xgb_pred, target_names=label_encoder.classes_))
print("XGBoost Accuracy:", accuracy_score(y_test, xgb_pred))
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
import numpy as np, json, os

# 5-fold CV, stratified for classification
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_rf  = cross_val_score(rf,  X_train, y_train, cv=cv, scoring="accuracy")
cv_xgb = cross_val_score(xgb, X_train, y_train, cv=cv, scoring="accuracy")

print("RF  CV accuracy mean±std:", cv_rf.mean(), cv_rf.std())
print("XGB CV accuracy mean±std:", cv_xgb.mean(), cv_xgb.std())

os.makedirs("outputs", exist_ok=True)
with open("outputs/cv_summary.json", "w") as f:
    json.dump({
        "rf":  {"mean": float(cv_rf.mean()),  "std": float(cv_rf.std())},
        "xgb": {"mean": float(cv_xgb.mean()), "std": float(cv_xgb.std())}
    }, f, indent=2)
print("Saved → outputs/cv_summary.json")
```

```python
from sklearn.inspection import permutation_importance
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

os.makedirs("outputs", exist_ok=True)

def save_permutation_importance(model, X, y, name: str, top_n: int = 5):
    """
    Compute permutation importance on (X, y), save a CSV and a PNG bar chart.
    - name: logical model name, e.g. 'random_forest' or 'xgboost'
    - top_n: number of features to show
    """
    r = permutation_importance(
        model, X, y,
        n_repeats=10,
        random_state=42,
        n_jobs=-1
    )
    # Align with feature names if available (X is a DataFrame)
    if hasattr(X, "columns"):
        feat_names = list(X.columns)
        # In case model transforms columns internally (rare here), clip to min length
        m = min(len(feat_names), r.importances_mean.shape[0])
        feat_names = feat_names[:m]
        importances = r.importances_mean[:m]
    else:
        feat_names = [f"f{i}" for i in range(r.importances_mean.shape[0])]
        importances = r.importances_mean

    pi_df = pd.DataFrame({"feature": feat_names, "perm_importance": importances}) \
              .sort_values("perm_importance", ascending=False)
    top = pi_df.head(top_n)

    csv_path = f"outputs/{name}_permutation_importance_top{top_n}.csv"
    top.to_csv(csv_path, index=False)

    # Plot (horizontal bar)
    plot_df = top.iloc[::-1]
    plt.figure()
    plt.barh(plot_df["feature"], plot_df["perm_importance"])
    plt.title(f"Permutation Importance (Top {top_n}) – {name}")
    plt.xlabel("Decrease in score when permuted")
    plt.ylabel("Feature")
    plt.tight_layout()
    png_path = f"outputs/{name}_permutation_importance_top{top_n}.png"
    plt.savefig(png_path, dpi=200)
    plt.show()

    print(f"Saved → {csv_path}")
    print(f"Saved → {png_path}")
    print("Saved Top-5 permutation importance for RF & XGB in outputs/")
```

```python

save_permutation_importance(rf,  X_test, y_test, name="random_forest", top_n=10)
save_permutation_importance(xgb, X_test, y_test, name="xgboost",      top_n=10)
```

```python
import os, json, joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

os.makedirs("models", exist_ok=True)
os.makedirs("outputs", exist_ok=True)

TOP_N = 5  # <-- changed from 25 to 5

def save_model(model, name: str):
    path = f"models/{name}.pkl"
    joblib.dump(model, path)
    print(f"[{name}] Saved model → {path}")
    return path

def save_metrics(y_true, y_pred, model, name: str):
    metrics = {
        "model": type(model).__name__,
        "accuracy": float(accuracy_score(y_true, y_pred)),
        "classification_report": classification_report(y_true, y_pred, output_dict=True)
    }
    path = f"outputs/{name}_metrics.json"
    with open(path, "w") as f:
        json.dump(metrics, f, indent=2)
    print(f"[{name}] Saved metrics → {path}")
    return path

def save_confusion_matrix(y_true, y_pred, name: str, labels=None):
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    plt.figure()
    plt.imshow(cm, interpolation="nearest", cmap=plt.cm.Blues)
    plt.title(f"Confusion Matrix ({name})")
    plt.colorbar()
    ticks = np.arange(cm.shape[0])
    if labels is None:
        labels = ticks
    plt.xticks(ticks, labels, rotation=45, ha="right")
    plt.yticks(ticks, labels)
    plt.xlabel("Predicted"); plt.ylabel("Actual")
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, int(cm[i, j]), ha="center", va="center", color="black")
    plt.tight_layout()
    path = f"outputs/{name}_confusion_matrix.png"
    plt.savefig(path, dpi=200)
    plt.show()
    print(f"[{name}] Saved confusion matrix → {path}")
    return path

def _xgb_importances(model):
    """Return a numpy array of importances for XGBoost model.
    Prefers sklearn API's feature_importances_. Falls back to booster 'gain' mapping."""
    # sklearn API
    imp = getattr(model, "feature_importances_", None)
    if imp is not None and len(imp) > 0:
        return np.array(imp)
    # raw booster mapping (may not align to X_train.columns if feature names not propagated)
    try:
        booster = model.get_booster()
        score_map = booster.get_score(importance_type="gain")
        # Ensure dense array in feature index order f0, f1, ...
        max_idx = -1
        idx_map = {}
        for k in score_map.keys():
            if k.startswith("f"):
                i = int(k[1:])
                idx_map[i] = score_map[k]
                max_idx = max(max_idx, i)
        dense = np.zeros(max_idx + 1, dtype=float)
        for i, val in idx_map.items():
            dense[i] = val
        return dense
    except Exception:
        return None

def save_feature_importance(model, name: str, feature_names):
    # Try generic .feature_importances_ first (RF + many tree models)
    importances = getattr(model, "feature_importances_", None)

    # XGBoost special-case if needed
    if importances is None and "xgb" in name.lower():
        importances = _xgb_importances(model)

    if importances is None:
        raise ValueError(f"[{name}] Model does not expose feature importances.")

    # Align lengths if needed
    if len(importances) != len(feature_names):
        print(f"[{name}] WARNING: importance length ({len(importances)}) "
              f"!= feature_names length ({len(feature_names)}). "
              f"Proceeding with min-length match.")
        m = min(len(importances), len(feature_names))
        importances = np.asarray(importances)[:m]
        feature_names = list(feature_names)[:m]

    imp_df = pd.DataFrame({"feature": feature_names, "importance": importances})
    imp_df = imp_df.sort_values("importance", ascending=False)

    # Save top-N CSV
    top = imp_df.head(TOP_N)
    csv_path = f"outputs/{name}_feature_importance_top{TOP_N}.csv"
    top.to_csv(csv_path, index=False)

    # Plot top-N
    plot_df = top.iloc[::-1]  # reverse for horizontal bar
    plt.figure()
    plt.barh(plot_df["feature"], plot_df["importance"])
    plt.title(f"Top {TOP_N} Feature Importances ({name})")
    plt.xlabel("Importance"); plt.ylabel("Feature")
    plt.tight_layout()
    png_path = f"outputs/{name}_feature_importance_top{TOP_N}.png"
    plt.savefig(png_path, dpi=200)
    plt.show()

    print(f"[{name}] Saved feature importance → {png_path}")
    print(f"[{name}] Saved feature table → {csv_path}")
    return png_path, csv_path
```

```python
# Ensure we have a class label list for consistent CM axes
labels = sorted(np.unique(y_test))

# --- Random Forest ---
try:
    rf  # just to trigger NameError if not defined
    name = "random_forest"

    model_path = save_model(rf, name)

    y_pred_rf = rf.predict(X_test)
    metrics_path = save_metrics(y_test, y_pred_rf, rf, name)
    cm_path = save_confusion_matrix(y_test, y_pred_rf, name, labels=labels)

    # Feature names come from X_train.columns (Phase-2 feature set)
    fi_png, fi_csv = save_feature_importance(rf, name, feature_names=list(X_train.columns))

except NameError:
    print("[random_forest] Skipped: 'rf' model not found in this notebook.")
```

```python
# --- XGBoost ---
try:
    xgb  # trigger NameError if not defined
    name = "xgboost"

    model_path = save_model(xgb, name)

    y_pred_xgb = xgb.predict(X_test)
    metrics_path = save_metrics(y_test, y_pred_xgb, xgb, name)
    cm_path = save_confusion_matrix(y_test, y_pred_xgb, name, labels=labels)

    fi_png, fi_csv = save_feature_importance(xgb, name, feature_names=list(X_train.columns))

except NameError:
    print("[xgboost] Skipped: 'xgb' model not found in this notebook.")
```
